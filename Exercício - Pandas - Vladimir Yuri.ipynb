## Pandas

A partir do dataset `movies.csv` utilize o Pandas para responder as seguintes perguntas:

### Qual o gênero de filme mais lucrativo em todo o mundo?

```python
# ADICIONE O SEU CÓDIGO AQUI
import pandas as pd

movies = pd.read_csv("C:/Users/vladimir/Desktop/icd-master/datasets/movies.csv")
movies

movies_genre = movies.groupby("Major Genre").mean()

movie_top1_worldwidegross = movies_genre.sort_values(by="Worldwide Gross", ascending=False).iloc[0]

movie_top1_worldwidegross

```

### Qual o diretor dirigiu mais filmes?

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
# ADICIONE O SEU CÓDIGO AQUI
movies["Collection"] = movies["Worldwide Gross"] / movies["Production Budget"]

movies_collection = movies.groupby("Title").mean()

movies_top5_Collection = movies_collection.sort_values(by="Collection", ascending=False).iloc[0:5]

movies_top5_Collection
```
